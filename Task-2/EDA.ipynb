# Removing the useless features and one hot encoding the discrete data

```python
df.drop(columns=["trans_date_trans_time","first", "last", "trans_num", "unix_time", "dob", "street", "city", "state", "lat", "long"], inplace=True)
df.info()
```

```python
le=LabelEncoder()

cat_column = df.select_dtypes(include=["object"]).columns.tolist()

for col in cat_column:
    df[col]=le.fit_transform(df[col])

df.info()
```

```python
fraud_rate = df.groupby("category")["is_fraud"].mean().sort_values(ascending=False)
fraud_rate
```
